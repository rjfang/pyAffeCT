```python
from pyAffeCT.datasets import WESAD

wesad = WESAD(os.path.abspath('/content/drive/MyDrive/02 Research/01 Affective Computing/03 Benchmark/pyAffeCT/data/WESAD'))

# view the subject id list
subject_list = wesad.subject_list

num_classes = 2

# prepare for the dataframe, set number of classess to be 3
df_train, df_test = wesad.prepare_flatten_dataframe(subject_included=subject_list[:3], split_by_subject=False, window_length=4,num_classes=num_classes)

train_loader, test_loader, input_shape = wesad.df_to_dataloader(df_train,df_test,num_classes, 4)
```
