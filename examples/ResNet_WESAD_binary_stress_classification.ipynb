# Dataset

```python
from pyAffeCT.datasets import WESAD

wesad = WESAD(os.path.abspath('/content/drive/MyDrive/02 Research/01 Affective Computing/03 Benchmark/pyAffeCT/data/02 WESAD/WESAD'))

# view the subject id list
subject_list = wesad.subject_list

# prepare for the dataframe, set number of classess to be 3
df_train, df_test = wesad.prepare_flatten_dataframe(subject_included=subject_list[:5], window_length=3)

train_loader, test_loader, input_shape = wesad.df_to_dataloader(df_train,df_test, 2, 20)
```

# Network

```python
from pyAffeCT.models import ResNet

resnet = ResNet(input_shape = input_shape, num_classes=2)

resnet.init_weights(method=None)
```

# Train

```python
resnet.train(train_loader, test_loader)
```

# Evaluate

```python
resnet.test_accuracy(test_loader)
```
